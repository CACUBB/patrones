# Data Retrieval


```python
from scipy.io import loadmat
from sklearn.model_selection import train_test_split

from mlxtend.feature_selection import SequentialFeatureSelector as sfs
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier

#####################################################################################

data = loadmat("set05-face-detection")
X = data["f"]
Y = data["d"]

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2, shuffle=False)
print(X.shape, Y.shape)
```

```python
#####################################################################################
# Clean data

import numpy as np
import pandas as pd

X_train = pd.DataFrame(data = X_train)
X_test = pd.DataFrame(data = X_test)

# Create correlation matrix
corr_matrix = X_train.corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Find features with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.999)]


# Drop features 
X_train = X_train.drop(X_train.columns[to_drop], axis=1).values
X_test = X_test.drop(X_test.columns[to_drop], axis=1).values

Y_train = Y_train.ravel()
Y_test = Y_test.ravel()

#####################################################################################
# train and test set


print("X test shape:", X_test.shape)

print("Y t shape:", Y_test.shape,"\n")

print("X train shape:", X_train.shape)

print("Y train shape:", Y_train.shape)

```

# Funciones Generales

```python
def get_patches( image, w ):
    
    patches = [None]*(w*w)
    
    height = len(image)
    width = len(image[0])
    
    d_height = round(height/w,0)
    d_width = round(width/w,0)

    # get w*w image patches indexes
    min_x = [0]*(w*w)
    max_x = [0]*(w*w)
    min_y = [0]*(w*w)
    max_y = [0]*(w*w)
    
    for x in range(w):
        for y in range(w):

            pos = int(w*x + y)
            
            # Get patch boundaries
            min_x[pos] = int(d_width*x)
            max_x[pos] = int(min_x[pos] + d_width)
            min_y[pos] = int(d_height*y)
            max_y[pos] = int(min_y[pos] + d_height)

            #cropped = image[x1:x2,y1:y2]
            cropped = image[min_y[pos]:max_y[pos] , min_x[pos]:max_x[pos]]
            patches[pos] = cropped

    return patches


from sklearn.metrics import accuracy_score
def performance_for_features(xtrain , ytrain, xtest , ytest):
    
    # train classifier
    knn = KNeighborsClassifier(n_neighbors = 1)
    knn.fit(xtrain, ytrain)

    # predict and evaluate performance
    prediction = knn.predict(xtest)
    
    return accuracy_score(ytest, prediction)
```

# Feature Selection

```python
#####################################################################################
# Scaling data

from sklearn import preprocessing

# scaler = preprocessing.StandardScaler().fit(X_train)
# X_train = scaler.transform(X_train)
# X_test = scaler.transform(X_test)

normalizer = preprocessing.Normalizer().fit(X_train)  # fit does nothing
X_train = normalizer.transform(X_train)
X_test = normalizer.transform(X_test)
```

```python
#####################################################################################
# Select 100 best features

# buid clasifier
knn = KNeighborsClassifier(n_neighbors = 1)
knn.fit(X_train, Y_train)

# Build step forward feature selection
my_sfs = sfs(knn,
             k_features = 100,
             forward = True,
             floating = False,
             verbose = 2,
             scoring = 'accuracy',
             cv = 2)

# Perform SFFS
my_sfs = my_sfs.fit(X_train, Y_train.ravel())
    
# Which features?
feat_cols = list(my_sfs.k_feature_idx_)
print("Selected features list:",feat_cols)

# selected feats #20 [4, 88, 154, 155, 156, 157, 158, 159, 231, 244, 272, 300, 328, 381, 382, 383, 384, 385, 1083, 1115]
# selected feats #50 [4, 88, 154, 155, 156, 157, 158, 159, 231, 244, 272, 300, 328, 356, 381, 382, 383, 384, 385, 386, 458, 608, 609, 610, 611, 612, 613, 685, 835, 836, 837, 838, 839, 840, 912, 915, 929, 943, 957, 971, 1061, 1062, 1063, 1064, 1065, 1066, 1067, 1083, 1115, 1139]
```

# PCA

```python
# Build full model with selected features
X_train_new = X_train[:, feat_cols]
X_test_new = X_test[:, feat_cols]

knn_new = KNeighborsClassifier(n_neighbors = 1)
knn_new.fit(X_train_new, Y_train)


acc = round(performance_for_features(X_train_new , Y_train , X_test_new , Y_test),2)


#####################################################################################
#  reduce dimensionality to 2 to plot classes

from sklearn.decomposition import PCA
import pandas as pd

pca = PCA(n_components = 10)

X_train_new = pca.fit_transform(X_train_new)
X_test_new = pca.fit_transform(X_test_new)
```
